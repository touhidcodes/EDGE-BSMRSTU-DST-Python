```python
# check prime number with functions
def checkPrime():
    count = 0
    num = int(input("Enter your number:"))
 
    for i in range(2, num//2+1):
        if num % i == 0:
            count += 1
 
    if count == 0:
        print(num, "is a prime number")
    else:
        print(num, "is not a prime number")
    
checkPrime()     
```

```python
# equilateral and isolateral triangle

x,y,z = map(int, input("Enter three numbers: ").split())

def checkTriangle(a, b, c):
    if a==b and b==c and a==c:
        print("Triangle is equilateral")
    elif a==b or b==c or a==c:
        print("Triangle is isolateral")
    else:
        print("Other")

checkTriangle(x, y, z)
```

```python
# factorial

def factorial(num):
    count = 1
    for i in range(1, num+1):
        count = count * i
    print(count)

x =int(input("Enter your number: "))
factorial(x)
```

```python
# recursive factorial

def factorial(num):
    if num == 1:
        return 1
    return factorial(num-1) * num

x = int(input("Enter your number: "))
factorial(x)
```

```python
# recursive fibonacci series

def fibo(n):
    if n==1:
        return 1
    elif n==2:
        return 1
    else:
        return fibo(n-1)+fibo(n-2)

num = int(input("Enter your number: "))
print(num)
print(fibo(num))
```

```python
# decorators

def log_args(func):
    def wrapper(*args, **kwargs):
        print(f"Arguments passed to {func.__name__}: {args}, {kwargs}")
        return func(*args, **kwargs)
    return wrapper

@log_args
def  greet(name, age):
    print(f"Hello {name}, tou are {age} years old!")

greet("Alice", 30)
        
```
